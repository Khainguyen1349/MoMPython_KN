```python
import numpy as np
from scipy.spatial import Delaunay#, ConvexHull
import matplotlib.pyplot as plt
from meshlib import *
```

## Define a hull
Define the shape on the negative side of Y axis

```python
hull_bound_points = np.array([[10,-20],[-10,-20],[10,0],[-10,0]])
mesh_resolution = 2
tri = creatHull(hull_bound_points,mesh_resolution)
plt.triplot(tri.points[:,0], tri.points[:,1], tri.simplices)
plt.show()
```

# Hull extension

```python
hull_bound_points_ext = np.array([[10,10],[-10,10],[10,0],[-10,0]])
mesh_resolution_ext = 2
commonEdge = np.array([[10,0],[-10,0]])
tri_final = extentHull(tri,hull_bound_points_ext,mesh_resolution_ext,commonEdge)
# plot
plt.triplot(tri_final.points[:,0], tri_final.points[:,1], tri_final.simplices)
plt.show()
```

# How to create a loop

```python
hull_bound_points = np.array([[10,-10],[-10,-10],[10,-8],[-10,-8]])
mesh_resolution = 1
tri = creatHull(hull_bound_points,mesh_resolution)
plt.triplot(tri.points[:,0], tri.points[:,1], tri.simplices)
plt.show()
```

```python
hull_bound_points_ext = np.array([[-10,0],[-8,0],[-10,-8],[-8,-8]])
mesh_resolution_ext = 1
commonEdge = np.array([[-10,-8],[-8,-8]])
tri_final = extentHull(tri,hull_bound_points_ext,mesh_resolution_ext,commonEdge)
# plot
plt.triplot(tri_final.points[:,0], tri_final.points[:,1], tri_final.simplices)
plt.show()
```

```python
hull_bound_points_ext = np.array([[10,0],[8,0],[10,-8],[8,-8]])
mesh_resolution_ext = 1
commonEdge = np.array([[10,-8],[8,-8]])
tri_final = extentHull(tri_final,hull_bound_points_ext,mesh_resolution_ext,commonEdge)
# plot
plt.triplot(tri_final.points[:,0], tri_final.points[:,1], tri_final.simplices)
plt.show()
```

```python
hull_bound_points_ext = np.array([[-10,0],[10,0],[-10,2],[10,2]])
mesh_resolution_ext = 1
commonEdge1 = np.array([[-10,0],[-8,0]])
commonEdge2 = np.array([[10,0],[8,0]])
tri_final = closeLoopHull(tri_final,hull_bound_points_ext,mesh_resolution_ext,commonEdge1,commonEdge2)
# plot
plt.triplot(tri_final.points[:,0], tri_final.points[:,1], tri_final.simplices)
plt.show()
```
