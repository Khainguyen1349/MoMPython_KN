```python
import numpy as np
from scipy.spatial import Delaunay#, ConvexHull
import matplotlib.pyplot as plt
from meshlib import *
```

## Define a hull
Define the shape on the negative side of Y axis

```python
hull_bound_points = np.array([[-5,10],[-6,10],[-5,11],[-6,11]])*4
mesh_resolution = 4
tri = creatHull(hull_bound_points,mesh_resolution)
plt.triplot(tri.points[:,0], tri.points[:,1], tri.simplices)
plt.show()
```

```python
hull_bound_points_ext = np.array([[10,8],[-10,8],[10,10],[-10,10]])*4
mesh_resolution_ext = 2
commonEdge = np.array([[-5,10],[-6,10]])*4
tri = extentHull(tri,hull_bound_points_ext,mesh_resolution_ext,commonEdge)
plt.triplot(tri.points[:,0], tri.points[:,1], tri.simplices)
plt.show()
```
