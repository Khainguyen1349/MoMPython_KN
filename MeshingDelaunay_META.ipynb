```python
import numpy as np
from scipy.spatial import Delaunay#, ConvexHull
import matplotlib.pyplot as plt
from meshlib import *
```

## Define a hull
Define the shape on the negative side of Y axis

```python
N = 15
Nx = 2*N + 1
Ny = Nx
init_pos = [-N,-N]
posX, posY = np.meshgrid(np.arange(Nx)-N,np.arange(Ny)-N)
R = 0.05*(np.floor(np.sqrt(posX**2 + posY**2)%4)+1)
```

```python
hull_bound_points = np.array([[-1,-1],[1,-1],[-1,1],[1,1]])*2
mesh_resolution = 2
tri = creatHull(hull_bound_points,mesh_resolution)
tri.points = tri.points*R[0,0] + init_pos
plt.triplot(tri.points[:,0], tri.points[:,1], tri.simplices)
plt.show()
```

```python
for i in range(Nx):
    for j in range(Ny):
        if not((i==0) and (j==0)):
            tri_temp = creatHull(hull_bound_points,mesh_resolution)
            tri_temp.points = tri_temp.points*R[i,j] + [i-N,j-N]
            tri = addSeparatedShapes(tri, tri_temp)
```

```python
plt.triplot(tri.points[:,0], tri.points[:,1], tri.simplices)
plt.show()
```
