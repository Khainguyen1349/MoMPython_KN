```python
import numpy as np
from scipy.spatial import Delaunay#, ConvexHull
import matplotlib.pyplot as plt
from meshlib import *
```

## Define a hull
Define the shape on the negative side of Y axis

```python
hull_bound_points = np.array([[-1,-1],[1,-1],[-1,1],[1,1]])*2
mesh_resolution = 4
tri1 = creatHull(hull_bound_points,mesh_resolution)
tri1.points = tri1.points*0.5 + 10
plt.triplot(tri1.points[:,0], tri1.points[:,1], tri1.simplices)
plt.show()
```

```python
hull_bound_points = np.array([[-1,-1],[1,-1],[-1,1],[1,1]])*2
mesh_resolution = 4
tri2 = creatHull(hull_bound_points,mesh_resolution)
tri2.points = tri2.points*0.25 - 10
plt.triplot(tri2.points[:,0], tri2.points[:,1], tri2.simplices)
plt.show()
```

```python
tri = addSeparatedShapes(tri1, tri2)
```

```python
plt.triplot(tri.points[:,0], tri.points[:,1], tri.simplices)
plt.show()
```
